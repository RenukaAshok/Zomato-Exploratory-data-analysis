**USING CLASSIFICATION ALGORITHMS**

```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Step 2: Load the dataset
file_path = r'C:\Users\sss\Downloads\zomato.csv (2).zip'
data = pd.read_csv(file_path, encoding='latin-1')

# Step 3: Display the first few rows of the dataset to understand its structure
print(data.head())

# Step 4: Check for missing values
print("Missing values per column:\n", data.isnull().sum())

# Step 5: Handle missing values (if needed)
# Example: Dropping columns with many missing values or filling with appropriate values
data = data.dropna()  # This is a simple approach; we can refine based on the data

# Step 6: Encode categorical variables
label_encoder = LabelEncoder()
data['Cuisines'] = label_encoder.fit_transform(data['Cuisines'].astype(str))
data['City'] = label_encoder.fit_transform(data['City'].astype(str))
data['Locality'] = label_encoder.fit_transform(data['Locality'].astype(str))
data['Has Table booking'] = label_encoder.fit_transform(data['Has Table booking'])
data['Has Online delivery'] = label_encoder.fit_transform(data['Has Online delivery'])

# Step 7: Define the target variable
# Creating a binary classification target based on Aggregate Rating
data['High Review'] = (data['Aggregate rating'] >= 4.0).astype(int)

# Step 8: Split the data into features and target
X = data[['Votes', 'Average Cost for two', 'Cuisines', 'Price range', 
          'Has Table booking', 'Has Online delivery', 'City', 'Locality']]
y = data['High Review']

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Data preprocessing complete. Ready for model training.")
```

**LOGISTIC REGRESSION**

```python
# Step 1: Import the necessary libraries for Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Initialize the Logistic Regression model
log_reg_model = LogisticRegression(max_iter=1000)

# Step 3: Train the model on the training data
log_reg_model.fit(X_train, y_train)

# Step 4: Make predictions on the test data
y_pred = log_reg_model.predict(X_test)

# Step 5: Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Logistic Regression Accuracy: {accuracy:.10f}")

# Generate and print classification report
report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", report)

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Logistic Regression')
plt.show()

```

**RANDOM FOREST**

```python
# Step 1: Import necessary libraries for Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Initialize the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Step 3: Train the Random Forest model on the training data
rf_model.fit(X_train, y_train)

# Step 4: Make predictions on the test data
y_pred_rf = rf_model.predict(X_test)

# Step 5: Evaluate the model's performance
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf:.10f}")

# Generate and print classification report
report_rf = classification_report(y_test, y_pred_rf)
print("\nClassification Report:")
print(report_rf)

# Generate confusion matrix
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Random Forest Classifier')
plt.show()
```

**Top cities to start the Restaurant**

```python
import pandas as pd
import zipfile

# Path to your zipped file
zip_file_path = r"C:\Users\sss\Downloads\zomato.csv (2).zip"

# Open the zip file and extract the CSV
with zipfile.ZipFile(zip_file_path, 'r') as z:
    # List all files in the zip
    file_names = z.namelist()
    
    # Assuming there's only one CSV file, we use the first name
    csv_file_name = file_names[0]  
    
    # Load the CSV file into a DataFrame with appropriate encoding
    with z.open(csv_file_name) as csv_file:
        df = pd.read_csv(csv_file, encoding='ISO-8859-1')

# Group data by 'City' and calculate average rating and total votes
city_analysis = df.groupby('City').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum'
}).reset_index()

# Filter cities with a minimum number of votes to ensure reliability (e.g., at least 500 votes)
filtered_cities = city_analysis[city_analysis['Votes'] > 500]

# Sort cities by average rating and votes
top_cities = filtered_cities.sort_values(by=['Aggregate rating', 'Votes'], ascending=[False, False])

# Display the top cities
print("Top Cities to Start a Restaurant Based on Customer Reviews:")
print(top_cities[['City', 'Aggregate rating', 'Votes']].head(10))  # Show top 10 cities
```

```python
import pandas as pd
import zipfile
import matplotlib.pyplot as plt
import seaborn as sns

# Path to your zipped file
zip_file_path = r"C:\Users\sss\Downloads\zomato.csv (2).zip"

# Open the zip file and extract the CSV
with zipfile.ZipFile(zip_file_path, 'r') as z:
    file_names = z.namelist()
    csv_file_name = file_names[0]  
    with z.open(csv_file_name) as csv_file:
        df = pd.read_csv(csv_file, encoding='ISO-8859-1')

# Group data by 'City' and calculate average rating and total votes
city_analysis = df.groupby('City').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum'
}).reset_index()

# Filter cities with a minimum number of votes to ensure reliability
filtered_cities = city_analysis[city_analysis['Votes'] > 500]

# Sort cities by average rating and votes
top_cities = filtered_cities.sort_values(by=['Aggregate rating', 'Votes'], ascending=[False, False])

# Visualization of the top cities
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='City', data=top_cities.head(10), palette='coolwarm')
plt.title('Top Cities to Start a Restaurant Based on Customer Reviews')
plt.xlabel('Average Aggregate Rating')
plt.ylabel('City')
plt.grid(axis='x')
plt.xlim(0, 5)  # Assuming ratings are out of 5
plt.show()
```

**Top Locations to start the Restaurant**

```python
# Group data by 'Locality' and calculate average rating and total votes
locality_analysis = df.groupby('Locality').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum'
}).reset_index()

# Filter localities with a minimum number of votes to ensure reliability (e.g., at least 100 votes)
filtered_localities = locality_analysis[locality_analysis['Votes'] > 100]

# Sort localities by average rating and votes
top_localities = filtered_localities.sort_values(by=['Aggregate rating', 'Votes'], ascending=[False, False])

# Display the top localities
print("Top Localities to Start a Restaurant Based on Customer Reviews:")
print(top_localities[['Locality', 'Aggregate rating', 'Votes']].head(10))  # Show top 10 localities
```

```python
# Import necessary libraries
import pandas as pd

# Load your dataset (assuming it's already loaded in 'df')
# Here, 'Locality', 'Aggregate rating', 'Votes', and 'Average Cost for two' are key columns

# Filter out restaurants with a decent number of votes to ensure reliability
vote_threshold = 100  # Adjust as needed to include popular restaurants
filtered_df = df[df['Votes'] > vote_threshold]

# Sort by 'Aggregate rating' (descending) and 'Average Cost for two' (ascending) for affordability
affordable_nice_food = filtered_df.sort_values(by=['Aggregate rating', 'Average Cost for two'], ascending=[False, True])

# Select the top 10 cheapest and highly rated restaurants and only display Locality and Aggregate rating
top_affordable_places = affordable_nice_food[['Locality', 'Aggregate rating']].head(10)

# Display the results
print("Top Locations for Nice Food at an Affordable Cost:")
print(top_affordable_places)
```

```python
# Set up the figure size
plt.figure(figsize=(12, 6))

# Plot the stacked bar chart
top_affordable_places.set_index('Restaurant Name')[['Average Cost for two', 'Aggregate rating']].plot(
    kind='bar', 
    stacked=True, 
    colormap='viridis', 
    figsize=(12, 6)
)

# Add titles and labels
plt.title('Cost and Rating Comparison for Top Affordable Restaurants', fontsize=14)
plt.xlabel('Restaurant Name')
plt.ylabel('Value')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
